```python
# finding mode of the salaries
# without Statistics libraries

sal.sort()
chk=[] # Create an empty list to store the frequency of each salary

# Iterate through the salary list
i=0
while i<len(sal):
  chk.append(sal.count(sal[i]))  # Count the occurrences of the current salary in the list and append to list
  i+=1 # increment da counter

dic = dict(zip(sal, chk)) # Create a dictionary to map salaries to their frequencies
dic2 = {k for (k,v)in dic.items() if v== max(chk)} # Create a set to store salaries with the maximum frequency

print(dic2)
```

```python
# finding standard deviation of the salaries
# without statistics libraries

sdv = samvar**0.5 #XD
print(sdv)
```

# Mean, Median, Mode, Sample Variance & Sample Standard Deviation with Libraries

__________________________________

```python
from statistics import *
#import entirety of satistics lib
salm = salaries
```

```python
#calculate mean using Statistics Library
mn = mean(salm)

print(mn, meen)

if meen == mn:
  print('the same')
```

```python
#calculate median using Statistics Library
mdn = median(salm)

print(mdn, med)

if mdn == med:
  print('the same')
```

```python
#calculate mode using Statistics Library
md = mode(salm)

print(md, dic2)

#yes they are the same :D
```

```python
#calculate sample variance using Statistics Library
vari = variance(salaries)

print(vari, samvar)

if vari == samvar:
  print('the same')
```

```python
#calculate standard deviation using Statistics Library

stddv = stdev(salaries)
print(stddv, sdv)

if stddv == sdv:
  print('the same')
```

```python
# calculate range of the dataset

max(sal)-min(sal)
```

```python
# calculate coeff. variation interquart. range of the dataset

def calculate_cv_iqr(sal):

  # Calculate quartiles
  q1 = median(sal[:len(sal) // 2])
  q3 = median(sal[len(sal) // 2:])

  # Handle equal quartile case for IQR
  iqr = None
  if q1 != q3:
    iqr = q3 - q1

  return cv, iqr

# Calculate CV and IQR
cv, iqr = calculate_cv_iqr(sal)

# Print the results
print("Coefficient of Variation (CV):", cv)
print("Interquartile Range (IQR):", iqr)
```

```python
# calculate quar coeff of dispersion of the dataset

def calculate_qcd(sal):

  # Calculate quartiles
  q1 = median(sal[:len(sal) // 2])
  q3 = median(sal[len(sal) // 2:])

  # Calculate QCD
  qcd = (q3 - q1) / (2 * mdn)

  return qcd

# Calculate QCD
qcd = calculate_qcd(sal)

# Print the result
print("Quartile Coefficient of Dispersion (QCD):", qcd)
```
