if n = 12

image.png

-  2 components ✓
- verteces = edges ✓
- cut vertex on either ✓

```python
gra = {  "1" : {"2"},
         "2" : {"3", "4"},
         "3" : {"2", "4"},
         "4" : {"2", "3"},
         "5" : {"4"},
         "6" : {"3"},
         "a" : {"b"},
         "b" : {"c","d"},
         "c" : {"b", "d"},
         "d" : {"b", "c"},
         "e" : {"d"},
         "f" : {"c"}
         }

def uncon(n):

    if n < 12:
        print("No")
    else:
        print("Yes")

b = len(gra)
uncon(b)
```

∴ needed atleast 12 verteces/ nodes to satisfy the conditions and be considered as unconnected graph.
