```python
fpat = '/content/Valve_Player_Data.csv'

import pandas as p
import numpy as n
```

```python
steem = p.read_csv(fpat)
steem
```

```python
'''Total number of players (playing the top 100 games on Steam)'''

tps = n.sum(steem['Peak_Players'])
print('the total concurrent players playing the top 100 games in steam are: ',tps, 'players')
```

```python
'''Selected game's player data statistics'''
# For Grandt theft Auto V

gtav = p.DataFrame(steem)
gtav = steem[steem['Game_Name'] == 'Grand Theft Auto V'].copy()
gtav.drop(['URL'], axis = 1, inplace = True)
gtav
```

```python
gtav.describe()
```

```python
# GTA V

'''get the total players of the game based on peak pllayers'''

gtatot = n.sum(gtav['Peak_Players'])
print('Total players of GTA V in Steam from 2015 - 2021: ',gtatot, 'players')

'''lowest average players of the game'''

gtamin = n.min(gtav['Avg_players'])
print('Lowest players of GTA V: ',gtamin, 'players')


'''highest peak players of the game'''

hpgtav = n.max(gtav['Peak_Players'])
print('The highest peak players of GTA V: ',hpgtav, 'players')
```
