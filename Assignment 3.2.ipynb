```python
import random

print("╜φ°⌂▌╫§╜φ°⌂▌╫§╜φ°⌂▌╫§╜φ°⌂▌╫§╜φ°⌂▌╫§╜φ°⌂▌╫§╜φ°⌂▌\n")

class bilsplatt:
    def __init__(self):
        self.members_dict = {}
        self.v = 1

    def __str__(self):
        return f"Members: {self.members_dict}"

    def admem(self):

        n = int(input('number of members: '))
        for i in range(n):
            memName = input('name of member: ')
            if memName not in self.members_dict.values():
                self.members_dict[self.v] = [memName, 0]
                print('added')
            else:
                print('name already present.')
                continue
            self.v += 1
        print("All members have been added!")

    def transac(self):
        print("\n")
        amount_left = 0
        n = len(self.members_dict)
        m = int(input('number of transac: '))
        balance = [0 for i in range(n)]
        print(balance)
        for t in range(m):
            print('Transaction ID:', random.randint(1, 1000))
            np, ns = map(int, input('number of payers and the number of splits: ').split())

            for i in range(np):
                memid, amt = map(int, input('payer id & the amount of money you will lend: ').split())
                self.members_dict[memid][1] -= amt
                balance[memid - 1] -= amt
                amount_left -= amt

            a = ns
            for i in range(ns):
                print('You have', a, 'split/s left.')
                print('splittable money:', abs(amount_left))
                memid, amt = map(int, input('member id and the amount they will split: ').split())
                self.members_dict[memid][1] += amt
                balance[memid - 1] += amt
                a -= 1
                amount_left += amt

        temp = []
        j = 0
        for i in range(n):
            if balance[i] > 0:
                cur_bal = balance[i]
                while cur_bal > 0 and j < n:
                    if balance[j] >= 0:
                        j = j + 1
                        continue
                    minimum_bal = min(cur_bal, abs(balance[j]))
                    cur_bal -= minimum_bal
                    balance[j] += minimum_bal
                    temp.append([[i + 1, j + 1], minimum_bal])

        for i in range(len(temp)):
            print(temp[i][0][0], temp[i][0][1], temp[i][1])

    def showmem(self):
        print(self.members_dict)


xd = bilsplatt()
xd.admem()
xd.transac()
xd.showmem()

print("\n╜φ°⌂▌╫§╜φ°⌂▌╫§╜φ°⌂▌╫§╜φ°⌂▌╫§╜φ°⌂▌╫§╜φ°⌂▌╫§╜φ°⌂▌")
```
