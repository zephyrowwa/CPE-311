```python
fpat = '/content/Valve_Player_Data.csv'

import pandas as p
import numpy as n
```

```python
steem = p.read_csv(fpat)
steem
```

```python
'''remove the null values in my dataset
   (usually the first days of a game)'''

clsteem = steem.dropna()
clsteem
```

```python
'''Total number of players (playing the top 100 games on Steam)'''

tps = n.sum(clsteem['Peak_Players'])
print('the total concurrent players playing the top 100 games in steam are: ',tps, 'players\n')
mnstm = n.mean(clsteem['Avg_players'])
print('the total average players playing from 2016 -2021: ',round(mnstm, 2), 'players\n')


'''Date with the highest players'''

print("Date and Game with the highest players:")

# Find the index of the row with the highest peak players
mxpidx = clsteem['Peak_Players'].idxmax()
mxp = clsteem.iloc[[mxpidx]]
mxp
```

```python
'''analyze player growth'''

pgd = clsteem['Avg_players'].diff()
print("Monthly player change:\n", pgd)
```

```python
'''Selected game's player data statistics'''
# For CSGO

csgo = p.DataFrame(clsteem)
csgo = clsteem[clsteem['Game_Name'] == 'Counter Strike: Global Offensive'].copy()
csgo.drop(['URL'], axis = 1, inplace = True)
csgo
```

```python
# CSGO

'''get the total players of the game based on peak pllayers'''
csgotot = n.sum(csgo['Peak_Players'])
print('Total players of CS:GO in Steam from 2012 - 2021: ',csgotot, 'players\n')

'''mean of players of the game'''
csgomn = n.mean(csgo['Avg_players'])
print('Total average of CS:GO players: ',round(csgomn, 2), 'players')

'''lowest average players of the game'''
csgomin = n.min(csgo['Avg_players'])
print('Lowest players of CS:GO: ',csgomin, 'players')

'''highest peak players of the game'''
hpcsgo = n.max(csgo['Peak_Players'])
print('The highest peak players of CS:GO: ',hpcsgo, 'players')
```
